```python
# compute f1 score
def compF1(rec, pre):
    return 2 * (pre*rec) / (pre+rec)
```

```python
### SPACE
Rec_0 = 0.9798; Prec_0 = 0.9918
print(compF1(Rec_0, Prec_0))
```

```python
### COMMA
Rec_1 = 0.7602; Prec_1 = 0.6677
print(compF1(Rec_1, Prec_1))
```

```python
### PERIOD
Rec_2 = 0.7881; Prec_2 = 0.8480
print(compF1(Rec_2, Prec_2))
```

```python
### QUESTION
Rec_3 = 0.5652; Prec_3 = 0.7879
print(compF1(Rec_3, Prec_3))
```

```python
def compPerc(a, b):
    return (b - a) / a * 100  
```

```python
a = 0.667; b = 0.658
compPerc(a, b)
```
